```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
import os
from pathlib import Path

def analyze_face(image_path):
    """–ê–Ω–∞–ª–∏–∑ –ª–∏—Ü–∞ —Å –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –ø–æ–ª–∞ –∏ –≤–æ–∑—Ä–∞—Å—Ç–∞"""

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏—è —Ñ–∞–π–ª–∞
    if not os.path.exists(image_path):
        print(f"‚ùå –§–∞–π–ª {image_path} –Ω–µ –Ω–∞–π–¥–µ–Ω!")
        return

    # –ó–∞–≥—Ä—É–∑–∫–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è —Å –ø—Ä–æ–≤–µ—Ä–∫–æ–π
    image = cv2.imread(image_path)
    if image is None:
        print(f"‚ùå –û—à–∏–±–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è {image_path}")
        print("–ü—Ä–æ–≤–µ—Ä—å—Ç–µ:")
        print("1. –°—É—â–µ—Å—Ç–≤—É–µ—Ç –ª–∏ —Ñ–∞–π–ª")
        print("2. –ö–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –ª–∏ —Ñ–æ—Ä–º–∞—Ç —Ñ–∞–π–ª–∞")
        print("3. –ù–µ –ø–æ–≤—Ä–µ–∂–¥–µ–Ω –ª–∏ —Ñ–∞–π–ª")
        return

    print(f"‚úÖ –ò–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∑–∞–≥—Ä—É–∂–µ–Ω–æ —É—Å–ø–µ—à–Ω–æ")
    print(f"üìê –†–∞–∑–º–µ—Ä: {image.shape[1]}x{image.shape[0]} –ø–∏–∫—Å–µ–ª–µ–π")

    # –ö–æ–Ω–≤–µ—Ä—Ç–∞—Ü–∏—è —Ü–≤–µ—Ç–æ–≤–æ–≥–æ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∏—Å—Ö–æ–¥–Ω–æ–≥–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(image_rgb)
    plt.axis('off')
    plt.title('–ò—Å—Ö–æ–¥–Ω–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ')

    # –ê–Ω–∞–ª–∏–∑ –ª–∏—Ü–∞ —Å –ø–æ–º–æ—â—å—é DeepFace
    try:
        print("üîç –ê–Ω–∞–ª–∏–∑ –ª–∏—Ü–∞...")
        analysis = DeepFace.analyze(img_path=image_path,
                                  actions=['gender', 'age', 'emotion', 'race'],
                                  enforce_detection=True)

        # –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
        print("\n" + "="*50)
        print("üéØ –†–ï–ó–£–õ–¨–¢–ê–¢–´ –ê–ù–ê–õ–ò–ó–ê –õ–ò–¶–ê")
        print("="*50)

        result = analysis[0]

        print(f"üë§ –ü–æ–ª: {result['dominant_gender']}")
        print(f"   üìä –ú—É–∂—á–∏–Ω–∞: {result['gender']['Man']:.1f}%")
        print(f"   üìä –ñ–µ–Ω—â–∏–Ω–∞: {result['gender']['Woman']:.1f}%")

        print(f"üéÇ –í–æ–∑—Ä–∞—Å—Ç: {result['age']} –ª–µ—Ç")

        print(f"üòä –≠–º–æ—Ü–∏–∏: {result['dominant_emotion']}")
        emotions = result['emotion']
        for emotion, score in emotions.items():
            print(f"   üìä {emotion}: {score:.1f}%")

        print(f"üë• –†–∞—Å–∞: {result['dominant_race']}")
        races = result['race']
        for race, score in races.items():
            print(f"   üìä {race}: {score:.1f}%")

        print(f"üìç –û–±–ª–∞—Å—Ç—å –ª–∏—Ü–∞: {result['region']}")

        # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å bounding box
        x = result['region']['x']
        y = result['region']['y']
        w = result['region']['w']
        h = result['region']['h']

        # –†–∏—Å—É–µ–º bounding box –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–∏
        image_with_box = image_rgb.copy()
        cv2.rectangle(image_with_box, (x, y), (x+w, y+h), (0, 255, 0), 3)

        # –î–æ–±–∞–≤–ª—è–µ–º —Ç–µ–∫—Å—Ç —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏
        label = f"{result['dominant_gender']}, {result['age']} –ª–µ—Ç"
        cv2.putText(image_with_box, label, (x, y-10),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

        # –û—Ç–æ–±—Ä–∞–∂–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
        plt.subplot(1, 2, 2)
        plt.imshow(image_with_box)
        plt.axis('off')
        plt.title('–†–µ–∑—É–ª—å—Ç–∞—Ç –∞–Ω–∞–ª–∏–∑–∞')

        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –∞–Ω–∞–ª–∏–∑–µ –ª–∏—Ü–∞: {e}")
        print("\n–í–æ–∑–º–æ–∂–Ω—ã–µ –ø—Ä–∏—á–∏–Ω—ã:")
        print("1. –õ–∏—Ü–æ –Ω–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–æ –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–∏")
        print("2. –ü—Ä–æ–±–ª–µ–º—ã —Å —É—Å—Ç–∞–Ω–æ–≤–∫–æ–π DeepFace")
        print("3. –ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è")

# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏
image_path = '/content/asd.jfif'
analyze_face(image_path)
```

```python
import cv2
import numpy as np
from mtcnn import MTCNN
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
import matplotlib.pyplot as plt

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –¥–µ—Ç–µ–∫—Ç–æ—Ä–∞ –ª–∏—Ü
detector = MTCNN()

def load_gender_age_models():
    """–ó–∞–≥—Ä—É–∑–∫–∞ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ–ª–∞ –∏ –≤–æ–∑—Ä–∞—Å—Ç–∞"""
    # –≠—Ç–∏ –º–æ–¥–µ–ª–∏ –Ω—É–∂–Ω–æ –ø—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ —Å–∫–∞—á–∞—Ç—å –∏–ª–∏ –æ–±—É—á–∏—Ç—å
    # –î–ª—è –ø—Ä–∏–º–µ—Ä–∞, –∑–¥–µ—Å—å –±—É–¥—É—Ç –∑–∞–≥–ª—É—à–∫–∏
    print("–ó–∞–≥—Ä—É–∂–∞–µ–º –º–æ–¥–µ–ª–∏...")
    # –í —Ä–µ–∞–ª—å–Ω–æ–º –∫–æ–¥–µ –∑–¥–µ—Å—å –±—É–¥–µ—Ç –∑–∞–≥—Ä—É–∑–∫–∞ –≤–∞—à–∏—Ö –º–æ–¥–µ–ª–µ–π
    return None, None

def predict_gender_age(face_roi):
    """–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –ø–æ–ª–∞ –∏ –≤–æ–∑—Ä–∞—Å—Ç–∞ –¥–ª—è –æ–¥–Ω–æ–≥–æ –ª–∏—Ü–∞"""
    # –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –ª–∏—Ü–∞
    face_roi = cv2.resize(face_roi, (224, 224))
    face_roi = preprocess_input(face_roi)
    face_roi = np.expand_dims(face_roi, axis=0)

    # –ó–¥–µ—Å—å –±—É–¥–µ—Ç –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –º–æ–¥–µ–ª–µ–π
    # –î–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ–º —Å–ª—É—á–∞–π–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
    gender = "–ú—É–∂—á–∏–Ω–∞" if np.random.random() > 0.5 else "–ñ–µ–Ω—â–∏–Ω–∞"
    age = np.random.randint(20, 60)
    confidence = np.random.uniform(0.7, 0.95)

    return gender, age, confidence

# –ó–∞–≥—Ä—É–∑–∫–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
image_path = '/content/asd.jfif'
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# –û–±–Ω–∞—Ä—É–∂–µ–Ω–∏–µ –ª–∏—Ü
print("üîç –û–±–Ω–∞—Ä—É–∂–µ–Ω–∏–µ –ª–∏—Ü...")
faces = detector.detect_faces(image_rgb)

if len(faces) == 0:
    print("–õ–∏—Ü–∞ –Ω–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω—ã!")
else:
    print(f"–ù–∞–π–¥–µ–Ω–æ –ª–∏—Ü: {len(faces)}")

    # –ó–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–µ–π (–≤ —Ä–µ–∞–ª—å–Ω–æ–º –ø—Ä–æ–µ–∫—Ç–µ —Ä–∞—Å–∫–æ–º–º–µ–Ω—Ç–∏—Ä—É–π—Ç–µ)
    # gender_model, age_model = load_gender_age_models()

    # –û–±—Ä–∞–±–æ—Ç–∫–∞ –∫–∞–∂–¥–æ–≥–æ –ª–∏—Ü–∞
    for i, face in enumerate(faces):
        # –ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –ª–∏—Ü–∞
        x, y, w, h = face['box']
        x, y = max(0, x), max(0, y)

        # –í—ã–¥–µ–ª–µ–Ω–∏–µ –æ–±–ª–∞—Å—Ç–∏ –ª–∏—Ü–∞
        face_roi = image_rgb[y:y+h, x:x+w]

        # –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –ø–æ–ª–∞ –∏ –≤–æ–∑—Ä–∞—Å—Ç–∞
        gender, age, confidence = predict_gender_age(face_roi)

        # –û—Ç—Ä–∏—Å–æ–≤–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
        color = (0, 255, 0) if gender == "–ú—É–∂—á–∏–Ω–∞" else (255, 0, 255)
        cv2.rectangle(image_rgb, (x, y), (x+w, y+h), color, 2)

        # –¢–µ–∫—Å—Ç —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏
        label = f"{gender}, {age} –ª–µ—Ç ({confidence:.1%})"
        cv2.putText(image_rgb, label, (x, y-10),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

        print(f"–õ–∏—Ü–æ {i+1}: {gender}, {age} –ª–µ—Ç (—É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å: {confidence:.1%})")

# –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
plt.figure(figsize=(12, 8))
plt.imshow(image_rgb)
plt.axis('off')
plt.title('–†–µ–∑—É–ª—å—Ç–∞—Ç—ã –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ–ª–∞ –∏ –≤–æ–∑—Ä–∞—Å—Ç–∞')
plt.show()
```
