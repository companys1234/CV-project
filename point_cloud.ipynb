```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Генерация случайного облака точек
points = np.random.rand(100, 3) * 10  # Массив размером 100x3 (X,Y,Z координаты)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Отображение облака точек
ax.scatter(points[:, 0], points[:, 1], points[:, 2], c='blue', marker='o')
plt.show()

```
